```python
#Library
import pandas as pd
import numpy as np
import tensorflow as tf
import sklearn
from keras import backend as K
from tensorflow import keras
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Input
from sklearn.model_selection import train_test_split
from sklearn.model_selection import ShuffleSplit
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn import linear_model
from matplotlib import style
import matplotlib.pyplot as pyplot

```

```python
# Memasukan Data
dataset = "https://github.com/Nicopratama01/cobaygy/blob/main/coba.xlsx?raw=true"
data = pd.read_excel(dataset)
data.head()
```

```python
# Menentukan Data Masukan
X = np.array(data.drop(columns=["Time","Force"]))
# Normalisasi Data Inputs Test
scaler = MinMaxScaler()
scaler.fit(data.drop(columns=["Time","Force"]))
X
# Mendifinisikan Data Output Test
y = np.array(data["Force"])
```

```python
#Pembuatan Model
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y, test_size=0.2)
sklearn.model_selection.train_test_split
linear = linear_model.LinearRegression()
linear.fit(X_train, y_train)
acc = linear.score(X_test,y_test)
print (acc)
```

```python
#Mendifinisikan Grafik
predictions = linear.predict(X_test)
for x in range(len(predictions)):
    print(predictions[x], X_test[x], y_test[x])
    
p = 'Displacement'
style.use("ggplot")
pyplot.scatter(data[p], data["Force"], color="black")

pyplot.xlabel (p)
pyplot.ylabel ("Force")
pyplot.show()

p = 'Velocity'
style.use("ggplot")
pyplot.scatter(data[p], data["Force"], color="black")

pyplot.xlabel (p)
pyplot.ylabel ("Force")
pyplot.show()

p = 'Current'
style.use("ggplot")
pyplot.scatter(data[p], data["Force"], color="black")

pyplot.xlabel (p)
pyplot.ylabel ("Force")
pyplot.show()
```
